```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np


def allign(file):
    data = fits.open(file)
    hdu = data[0]
    startt = hdu.header['CRVAL1']
    size = hdu.header['NAXIS1']
    prwto = hdu.header['CRPIX1']
    step = hdu.header['CDELT1']
    if prwto!=1.:
        prwto=prwto
    elif prwto==1:
        prwto=0
    size = size-prwto #give the right wavelength to the first pixel of the spectrum
    lamda = np.arange(startt,startt+size*step,step) #beginning to end of spectrum
    flux = hdu.data[prwto:]
    name = str(file[53:-4])
    data.close()
    return lamda,flux,name
```

```python
def plot_spec(lamda,flux,name):
    plt.figure(figsize=(16,6))
    fig = plt.plot(lamda,flux, '-k')
    plt.grid()
    plt.xticks(size=14)
    plt.yticks(size=14)
    plt.xlabel('Wavelength / '+r'$\mathrm{\AA}$',size=14)
    plt.title('HE'+name[0:9],size=19)
    plt.tight_layout()
    plt.savefig('Abbildungen/Spektren/HE'+name)
    return fig
```

```python
def plot_spec_zoom(lamda,flux,name,lmin,lmax):
    plt.figure(figsize=(16,6))
    fig = plt.plot(lamda,flux, '-k')
    plt.grid()
    plt.xlim(lmin,lmax)
    plt.xticks(size=14)
    plt.yticks(size=14)
    plt.xlabel('Wavelength / '+r'$\AA$',size=14)
    plt.title('HE'+name[0:9],size=19)
    plt.savefig('Abbildungen/Spektren/HE'+name)
    return fig
```
